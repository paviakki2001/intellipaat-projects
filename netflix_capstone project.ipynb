```python
# create variable for customer_count

cust_id_count=netflix_dataset['Cust_ID'].nunique()

customer_count=cust_id_count - movie_count # as the cust_id consist of movie_id therefore after removing it we will get actual customer_count

print( f" {'Customer_count : '}{customer_count} "  ) # total no. of customer
```

```python
# creating variable for rating_count

rating_count=netflix_dataset['Cust_ID'].count() - movie_count # as rating consist of NaN so removing it will give total rating count

print( f" {'Rating_count : '}{rating_count} " )
```

```python
# now create a dataframe for movie_id so that we can add new column in dataset as 'Movie_id'

counter=None
movie_np=[]
# Iterate over the DataFrame rows
for i in netflix_dataset['Cust_ID']:
     if ':' in i :
          # Update the current movie ID
          counter=int(i.replace(':',''))
     movie_np.append(counter)
```

```python
movie_np
```

```python
netflix_dataset['Movie_ID']=movie_np
```

```python
# as above new column is created for movie_id

netflix_dataset
```

```python
# removing NaN values as movie_id column consist the information

netflix_dataset=netflix_dataset[netflix_dataset['Rating'].notna()]

# notna() is used to remove NaN/null values
```

```python
netflix_dataset.info()
```

```python
# as Cust_id consist of object dtype changing it to int dtype before model building

netflix_dataset['Cust_ID']=netflix_dataset['Cust_ID'].astype(int)

netflix_dataset.info()
```

```python
# Pre filtering

# 1. filtering all those movies that have less rating
# 2. filtering all those users which have rated less movies
```

# 1. Filtering movies having Less ratings

```python
df_movie_summary=netflix_dataset.groupby('Movie_ID')['Movie_ID'].agg(['count'])
df_movie_summary
```

```python
# now taking the overall quantile of 0.60 for as a set benchmark

movie_benchmark=round(df_movie_summary['count'].quantile(0.60))
movie_benchmark
```

```python
# now creating a list of movies have less than benchmarks
movie_drop_list=df_movie_summary[df_movie_summary['count']<movie_benchmark].index
movie_drop_list
```

```python
# Actual movies left

Actual_movies = movie_count - len(movie_drop_list)
print ( f" {'Acutal_Movies_left : '}{Actual_movies} ")
```

# 2. Filtering users who rated less

```python
df_customer_summary=netflix_dataset.groupby('Cust_ID')['Cust_ID'].agg(['count'])
df_customer_summary
```

```python
# now taking the overall quantile of 0.60 for as a set benchmark

customer_benchmark=round(df_customer_summary['count'].quantile(0.60))
customer_benchmark
```

```python
# removing customer who doesn't meet benchmarks

customer_drop_list=df_customer_summary[df_customer_summary['count']<customer_benchmark].index
customer_drop_list
```

```python
# actual no. customer

Actual_customer=customer_count - len(customer_drop_list)
print( f" {'Actual_Customer_left : '}{Actual_customer}")
```

```python
# now removing movie_drop_list and customer_drop-list from original dataset

netflix_dataset=netflix_dataset[~netflix_dataset['Movie_ID'].isin(movie_drop_list)]
netflix_dataset=netflix_dataset[~netflix_dataset['Cust_ID'].isin(customer_drop_list)]
print(  f" {'After trimming the shape of Dataset left : '}{netflix_dataset.shape}")
```

```python
from surprise import Dataset, Reader, SVD
from surprise.model_selection import cross_validate
```

```python
reader=Reader()
```

```python
# for quickrun time using first 100k rows

data=Dataset.load_from_df(netflix_dataset[['Cust_ID','Movie_ID','Rating']][:100000],reader)
```

```python
# creating SVD model

model=SVD()
```

```python
cross_validate(model,data,measures=['RMSE'],cv=3)
```
